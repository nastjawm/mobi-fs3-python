### Not using the `ndarray.clip` method

**<a id="#">Task 2.3.2 (a).</a>** Enter your solution below:

```python
T1 = 0.31
T2 = 0.59
img = plt.imread('data/cells.png')
img_clip = img.copy()
seq = img.flatten()
rows = img.shape[0]
cols = img.shape [1]
for row in range(rows):
    for j in range(cols):
        if img[row, j] < T1:
            img_clip[row, j] = T1
        if img[row, j] > T2:
            img_clip[row, j] = T2
        else:
            img_clip[row, j] = img[row, j]
    
plt.imshow(img_clip, 'gray')
plt.colorbar()
```

**<a id="#">Task 2.3.2 (b).</a>** Enter your solution below:

```python
T1 = 0.31
T2 = 0.59
img = plt.imread('data/cells.png')
img_clip = img.copy()
for idx in numpy.ndindex(img.shape):
    row = idx[0]
    col = idx[1]
    if img[row, col] < T1:
        img_clip[row, col] = T1
    if img[row, col] > T2:
        img_clip[row, col] = T2
    else:
        img_clip[row, col] = img[row, col]

plt.imshow(img_clip, 'gray')
plt.colorbar()
```
